```python
from sklearn.datasets import load_iris 
from sklearn.decomposition import PCA

iris = load_iris()
X = iris.data
y = iris.target

pca_transformer = PCA(n_components = 1)
pca_iris = pca_transformer.fit(X, y)
print(pca_transformer.explained_variance_ratio_)

pca_transformer = PCA(n_components = 2)
pca_iris = pca_transformer.fit(X, y)
print(pca_transformer.explained_variance_ratio_)

pca_transformer = PCA(n_components = 3)
pca_iris = pca_transformer.fit(X, y)
print(pca_transformer.explained_variance_ratio_)

pca_transformer = PCA(n_components = 4)
pca_iris = pca_transformer.fit(X, y)
print(pca_transformer.explained_variance_ratio_)
```

```python
from sklearn import preprocessing
import numpy as np

scaler = preprocessing.StandardScaler().fit(X)
X_scaled = scaler.transform(X)
print (X_scaled.mean(axis=0))
print (X_scaled.std(axis=0))

pca_transformer = PCA(n_components = 1)
pca_iris = pca_transformer.fit(X_scaled, y)
print(pca_transformer.explained_variance_ratio_)

pca_transformer = PCA(n_components = 2)
pca_iris = pca_transformer.fit(X_scaled, y)
print(pca_transformer.explained_variance_ratio_)

pca_transformer = PCA(n_components = 3)
pca_iris = pca_transformer.fit(X_scaled, y)
print(pca_transformer.explained_variance_ratio_)

pca_transformer = PCA(n_components = 4)
pca_iris = pca_transformer.fit(X_scaled, y)
print(pca_transformer.explained_variance_ratio_)
```

```python
import matplotlib.pyplot as plt
pca = PCA(n_components=2)
question_3 = pca.fit(X_scaled).transform(X_scaled)
x_3 = question_3[:,0]
y_3 = question_3[:,1]
plt.scatter(x_3, y_3, c = y)
plt.show()
```

```python
from sklearn.cluster import KMeans
knn_pca = KMeans(n_clusters = 3, init = 'k-means++')
knn_pca.fit(question_3)
knn_x = question_3[:,0]
knn_y = question_3[:,1]
plt.scatter(knn_x, knn_y, c = knn_pca.labels_)
plt.show()
```
